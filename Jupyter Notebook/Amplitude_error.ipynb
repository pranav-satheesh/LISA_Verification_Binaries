```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
M0 = 1.989 * 10**30
pc = 3.086 * 10**16


m1 = 0.49 * M0
m2 = 0.048 * M0


D = 197 * pc
L = 2.5 * 10**9
P = 1102.70
i = 26
mu = (m1*m2)/(m1+m2);
G = 6.677 * 10**(-11) 
c = 3 * 10**8
R = P**(2/3) * (G*(m1+m2)/(4*np.pi**2))**(1/3)
omega = 2*np.pi/P;


def Dllplus(Psi):
    return (G * mu * R**2 * omega * np.cos(2*Psi) * (3 + np.cos(2*i))) /(c**3 * D * L)
            
def Dllcross(Psi):
    
    return (G * mu * R**2 * omega * np.sin(2*Psi) * np.cos(i)) /(c**3 * D * L)            
            
```

```python
psi = np.linspace(0,np.pi/4,50)
```

```python
plt.figure(figsize=(10,7))
plt.rcParams['font.size'] = 12

plt.plot(psi*180/np.pi,Dllplus(psi),label = r'$\frac{|\Delta L_{\plus}|}{L}$')
plt.plot(psi*180/np.pi,Dllcross(psi),"r",label = r'$\frac{|\Delta L_{x}|}{L}$')


plt.fill_between(psi*180/np.pi,Dllplus(psi)-Dllplus(psi)*sigma_plus(2*np.pi/180,psi), Dllplus(psi)+Dllplus(psi)*sigma_plus(2*np.pi/180,psi),alpha=0.2, facecolor='#1B2ACC')
plt.fill_between(psi*180/np.pi,Dllplus(psi)-Dllplus(psi)*sigma_plus(5*np.pi/180,psi), Dllplus(psi)+Dllplus(psi)*sigma_plus(5*np.pi/180,psi),alpha=0.2, facecolor='#1B9555')
plt.fill_between(psi*180/np.pi,Dllplus(psi)-Dllplus(psi)*sigma_plus(10*np.pi/180,psi), Dllplus(psi)+Dllplus(psi)*sigma_plus(10*np.pi/180,psi),alpha=0.2, facecolor='#1B9748')

plt.fill_between(psi*180/np.pi,Dllcross(psi)-Dllcross(psi)*sigma_cross(2*np.pi/180,psi), Dllcross(psi)+Dllcross(psi)*sigma_cross(2*np.pi/180,psi),alpha=0.2, facecolor='#D62ACC')
plt.fill_between(psi*180/np.pi,Dllcross(psi)-Dllcross(psi)*sigma_cross(5*np.pi/180,psi), Dllcross(psi)+Dllcross(psi)*sigma_cross(5*np.pi/180,psi),alpha=0.2, facecolor='#D79555')
plt.fill_between(psi*180/np.pi,Dllcross(psi)-Dllcross(psi)*sigma_cross(10*np.pi/180,psi), Dllcross(psi)+Dllcross(psi)*sigma_cross(10*np.pi/180,psi),alpha=0.2, facecolor='#D99748')

plt.ylim(0,5*1e-21)
plt.xlim(0,45)
plt.legend()

plt.xlabel(r'$\Psi[^{\circ}]$')
plt.ylabel("Amplitude")
plt.savefig("LISA_ampitude_variation.jpg")
```

```python
def sigma_plus(dpsi,psi):
    return 2*np.sqrt(0.02**2 + dpsi**2 * np.tan(2*psi)**2)
```

```python
def sigma_cross(dpsi,psi):
    return 2*np.sqrt(0.03**2 + dpsi**2 * (1/np.tan(2*psi))**2)
```
