```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def response2(i,psi):
    i = np.pi/180 * i
    psi = np.pi/180 * psi
    csi = np.cos(i)
    cs = (1+csi**2) * np.cos(2*psi)
    sn = 2*csi * np.sin(2*psi)
    return (cs**2 + sn**2)**.5
```

```python
plt.figure(figsize=(16,9))
psi = np.linspace(0,90,100)
plt.plot(psi,response2(0,psi),"r-",label="$i=0^\circ$")
plt.plot(psi,response2(30,psi),"k-",label="$i=30^\circ$")
plt.plot(psi,response2(45,psi),"c-",label="$i=45^\circ$")
plt.plot(psi,response2(60,psi),"g-",label="$i=60^\circ$")
plt.plot(psi,response2(90,psi),"b-",label="$i=90^\circ$")
plt.ylim(0,2.2)
plt.legend()
plt.xlabel("Orientation (degrees)")
plt.ylabel("$\delta l$")
plt.title("$\delta l$ vs orientation for varying inclinations")
plt.show()
```
